````python
#CHALLENGE 2

#create two dictionaries, one holding latitudes and one holding longitudes.
lat = {} #Defining dictionary for latitudes
long = {} #Defining dictionary for longitudes
f = open("airports.dat.txt") #Importing data
for row in csv.reader(f):
  airport_id = row[0]
  lat[airport_id] = float(row[6])
  long[airport_id] = float(row[7])
lat
long
````

````python
#CHALLENGE 3

#Calculating geographic distances
# Copy of geo_distance.py
# Using the Haversine formula for geographic Great Circle Distance
# As per https://en.wikipedia.org/wiki/Haversine_formula

from math import cos,radians,sin,pow,asin,sqrt

def distance(lat1, long1, lat2, long2):
  radius = 6371 # radius of the earth in km, roughly https://en.wikipedia.org/wiki/Earth_radius
  # Lat,long are in degrees but we need radians
  lat1 = radians(lat1)
  lat2 = radians(lat2)
  long1 = radians(long1)
  long2 = radians(long2)

  dlat = lat2-lat1
  dlon = long2-long1

  a = pow(sin(dlat/2),2) + cos(lat1)*cos(lat2)*pow(sin(dlon/2),2)
  distance = 2 * radius * asin(sqrt(a))

  return distance
````

````python
#build list of route distances
import csv 
distances=[] 
f = open("routes.dat.txt") #importing data
for row in csv.reader(f): 
  source_airport= row[3] #defining source 
  destination_airport=row[5] #defining destination
  if source_airport in lat and destination_airport in lat: 
    #define source latitudes and longitudes from the previously defined dictionary 
    source_lat=lat[source_airport] 
    source_long=long[source_airport] 
    #define destination latitudes and longitudes from previously defined dictionary
    destination_lat=lat[destination_airport] 
    destination_long=long[destination_airport]
    distances.append(distance(source_lat,source_long,destination_lat,destination_long)) 
distances
````

````python
#CHALLENGE 4

#create a histogram displaying the frequency of flights by distance
import numpy as np
import matplotlib.pyplot as plt
plt.hist(distances,100,facecolor='purple')
plt.xlabel("distance in km")
plt.ylabel("number of flights")
````
